```python
import pandas as pd
df = pd.read_csv('data1.csv')
```

Menampilkan nama kolom dan tipe datanya

```python
df.info()
```

Menampilkan data pada data frame

Menampilkan jumlah data kosong pada kolom dan persentasenya

```python
missing_data = df.isnull().sum()
missing_data
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
missing_percentage = (missing_data / len(df)) * 100
missing_summary = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percentage})
missing_summary
```

Menghapus kolom yang memiliki jumlah data kosong lebih dari 50%

```python
columns_to_drop = missing_percentage[missing_percentage >= 50].index
df_cleaned = df.drop(columns=columns_to_drop)
```

Menampilkan kembali jumlah data kosong

```python
missing_data = df_cleaned.isnull().sum()
missing_data
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
missing_percentage = (missing_data / len(df_cleaned)) * 100
missing_summary = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percentage})
missing_summary
```

```python
df_cleaned.tail()
```

Menghapus baris yang memiliki data kosong lebih dari 50%

```python
threshold = len(df_cleaned.columns) / 2
df_cleaned = df_cleaned.dropna(thresh=threshold)
df_cleaned
```

Menampilkan kembali jumlah data kosong

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned["Jurusan / Prodi"].unique()
```

Menghapus kolom yang tidak diperlukan

```python
df_cleaned = df_cleaned.drop(columns=['Email Address', 'Kelas', 'Nama Peminjam', 'Tanggal', 'waktu', 'No HP WA', 'KPM / KTP / SIM'])
df_cleaned.head()
```

Mengubah tipe data kolom Timestamp dari object menjadi datetimes

```python
df_cleaned['Timestamp'] = pd.to_datetime(df_cleaned['Timestamp'])
```

```python
df_cleaned.info()
```

Membuat kolom baru yaitu Hari berdasarkan nilai Timestamp

```python
df_cleaned['Hari'] = df_cleaned['Timestamp'].dt.day_name()
```

```python
df_cleaned['Hari'] = df_cleaned['Hari'].map({'Sunday': 'Minggu', 'Monday': 'Senin', 'Tuesday': 'Selasa', 'Wednesday': 'Rabu', 'Thursday': 'Kamis', 'Friday': 'Jumat', 'Saturday': 'Sabtu'})
df_cleaned.tail()
```

```python
print(df_cleaned['Hari'].unique())
```

Membuat kolom baru gabungan Jurusan / Prodi dan Hari

```python
df_cleaned['Jurusan_Hari'] = df_cleaned['Jurusan / Prodi'] + ' - ' + df_cleaned['Hari']
```

```python
df_cleaned.info()
```

Mengibah format data menjadi daftar transaksi

```python
transactions = df_cleaned[['Jurusan_Hari', 'Nama Laboratorium']].values.tolist()
```

Mengubah daftar transaksi menjadi format one-hot encoding

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

Menjalankan Algoritma apriori untuk mencari frekuensi data

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_encoded, min_support=0.001, use_colnames=True)
```

Membuat aturan asosiasi berdasarkan itemset yang sering muncul

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)
```

Mengecek apakah rules kosong

```python
print(rules.empty)
```

Menampilkan bebrapa fitur pada rules

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

Menampilkan hasil dalam bentuk kalimat

```python
for index, row in rules.iterrows():
    print(f"Jika {set(row['antecedents'])} maka {set(row['consequents'])} dengan confidence {row['confidence']:.2f} dan lift {row['lift']:.2f}")
```
